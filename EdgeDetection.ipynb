```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import math

size = (557, 417)
img = cv2.imread('building.tif', 0)
img = cv2.resize(img, size)
print('Original Image: ')
cv2_imshow(img)

#defining sobel operator
sobel_x = np.zeros([3,3])
sobel_y = np.zeros([3,3])

sobel_x[0,0] = -1
sobel_x[1,0] = -2
sobel_x[2,0] = -1
sobel_x[0,1] = 0
sobel_x[1,1] = 0
sobel_x[2,1] = 0
sobel_x[0,2] = 1
sobel_x[1,2] = 2
sobel_x[2,2] = 1

sobel_y[0,0] = -1
sobel_y[1,0] = 0
sobel_y[2,0] = 1
sobel_y[0,1] = -2
sobel_y[1,1] = 0
sobel_y[2,1] = 2
sobel_y[0,2] = -1
sobel_y[1,2] = 0
sobel_y[2,2] = -1

#applying Gaussian smoothing
img = cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT)
#print('\nImage after smoothing: ')
#cv2_imshow(img)

#applying kernel
Gx = cv2.filter2D(img,-1,sobel_x)
Gy = cv2.filter2D(img,-1,sobel_y)

G = np.zeros([img.shape[0], img.shape[1]])
#combining result
for x in range(0, img.shape[0]):
  for y in range(0, img.shape[1]):
    G[x,y] = math.sqrt(math.pow(Gx[x,y], 2) + math.pow(Gy[x,y], 2))
print('\nApplying sobel kernel: ')
cv2_imshow(G)
```

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import math

size = (557, 417)
img = cv2.imread('building.tif', 0)
img = cv2.resize(img, size)
print('Original Image: ')
cv2_imshow(img)

#defining sobel operator
prewitt_x = np.zeros([3,3])
prewitt_y = np.zeros([3,3])

prewitt_x[0,0] = 1
prewitt_x[1,0] = 1
prewitt_x[2,0] = 1
prewitt_x[0,1] = 0
prewitt_x[1,1] = 0
prewitt_x[2,1] = 0
prewitt_x[0,2] = -1
prewitt_x[1,2] = -1
prewitt_x[2,2] = -1

prewitt_y[0,0] = 1
prewitt_y[1,0] = 0
prewitt_y[2,0] = -1
prewitt_y[0,1] = 1
prewitt_y[1,1] = 0
prewitt_y[2,1] = -1
prewitt_y[0,2] = 1
prewitt_y[1,2] = 0
prewitt_y[2,2] = -1

#applying Gaussian smoothing
#img = cv2.GaussianBlur(img,(9,9),cv2.BORDER_DEFAULT)
#print('\nImage after smoothing: ')
#cv2_imshow(img)

#applying kernel
Gx = cv2.filter2D(img,-1,prewitt_x)
Gy = cv2.filter2D(img,-1,prewitt_y)

G = np.zeros([img.shape[0], img.shape[1]])
#combining result
for x in range(0, img.shape[0]):
  for y in range(0, img.shape[1]):
    G[x,y] = math.sqrt(math.pow(Gx[x,y], 2) + math.pow(Gy[x,y], 2))
print('\nApplying prewitt kernel: ')
cv2_imshow(G)
```
