```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt

def smoothing(img):
  height = int(img.shape[0])
  width = int(img.shape[1])
  #create kernel
  w = np.zeros([3,3])
  w[0,0] = 1
  w[1,0] = 2
  w[2,0] = 1
  w[0,1] = 2
  w[1,1] = 4
  w[2,1] = 2
  w[0,2] = 1
  w[1,2] = 2
  w[2,2] = 1
  w = w * 1/16

  #zero-padding the image
  #img = cv2.copyMakeBorder(img, 1, 1, 1, 1, cv2.BORDER_CONSTANT)

  #applying correlation
  for x in range(0,(height-2)):
    for y in range(0,(width-2)):
      img[x+1][y+1] = w[0][0]*img[x][y] + w[0][1]*img[x+1][y] + w[0][2]*img[x+2][y] + w[1][0]*img[x][y+1] + w[1][1]*img[x+1][y+1] + w[1][2]*img[x+2][y+1] + w[2][0]*img[x][y+2] + w[2][1]*img[x+1][y+2] + w[2][2]*img[x+2][y+2]
  return img

image = cv2.imread('moon.tif',0)
#image = cv2.copyMakeBorder(image, 1, 1, 1, 1, cv2.BORDER_CONSTANT)
k = 0.5
print('Original Image: ')
cv2_imshow(image)
print('\n')
#call smoothing function
smooth = cv2.GaussianBlur(image,(5,5),cv2.BORDER_DEFAULT)
#smooth = smoothing(image)
print("Image after Gaussian blur: ")
cv2_imshow(smooth)
mask = image - smooth
#high boosting
boosted = image + (0.5 * mask)
print("High boosted image:")
cv2_imshow(boosted)
```

```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt

img = cv2.imread('blade.tif', 0)
print("original Image: ")
cv2_imshow(img)
height = int(img.shape[0])
width = int(img.shape[1])
#create kernel
w = np.zeros([3,3])
w[0,0] = -1
w[1,0] = -1
w[2,0] = -1
w[0,1] = 2
w[1,1] = 2
w[2,1] = 2
w[0,2] = -1
w[1,2] = -1
w[2,2] = -1
#print(w)
#zero-padding the image
#img = cv2.copyMakeBorder(img, 1, 1, 1, 1, cv2.BORDER_CONSTANT)
"""
applying correlation
for x in range(0,(height-2)):
  for y in range(0,(width-2)):
    img[x+1][y+1] = w[0][0]*img[x][y] + w[0][1]*img[x+1][y] + w[0][2]*img[x+2][y] + w[1][0]*img[x][y+1] + w[1][1]*img[x+1][y+1] + w[1][2]*img[x+2][y+1] + w[2][0]*img[x][y+2] + w[2][1]*img[x+1][y+2] + w[2][2]*img[x+2][y+2]
"""
#after smoothing using laplacian kernel
res = cv2.filter2D(img,-1, w)
print("\nSmoothed Image")
cv2_imshow(res)

#applying threshold
for x in range(0, height):
  for y in range(0, width):
    if res[x,y] < 200:
      res[x,y] = 0
    else:
      res[x,y] = 255
print("\nImage after thresholding:")
cv2_imshow(res)
      
```
