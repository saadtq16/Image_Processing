```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('lena.tiff', 0)
print('Image before down-scaling:')
cv2_imshow(img)
h = int(img.shape[0])
w = int(img.shape[1])
new_h = h * 2
new_w = w * 2

i = 0
j = 0

new_img = np.zeros([new_h,new_w])

for y in range(0,h-1):
  for x in range(0,w-1):
    fa = ((((x+1)-(x+0.5))/((x+1)-x))*int(img[x][y]))+((((x+0.5)-x)/((x+1)-x))*int(img[x+1][y]))
    fb = ((((x+1)-(x+0.5))/((x+1)-x))*int(img[x][y+1]))+((((x+0.5)-x)/((x+1)-x))*int(img[x+1][y+1]))
    new_pix = ((((y+1)-(y+0.5))/((y+1)-y))*fa)+((((y+0.5)-y)/((y+1)-y))*fb)
    new_img[i][j] = int(img[x][y])
    new_img[i+1][j] = new_pix
    new_img[i][j+1] = new_pix
    new_img[i+1][j+1] = new_pix
    i += 2
  j += 2
  i = 0

print('Image after up-scaling:')
cv2_imshow(new_img)


```
